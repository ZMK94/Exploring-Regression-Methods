```python
import pandas as pd

air_foil = pd.read_table("airfoil_self_noise.dat", header= None, names=['Frequency', 'Angle','Chord_length',
                                                             'Velocity', 'Suction','Sound'])

air_foil.head()
```

## Check for missing values

```python
air_foil.isnull().sum(axis=0)
```

## Split the Data

```python
#obtain features and label
X = air_foil.drop('Sound',axis=1)
y = air_foil['Sound']

#split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)
```

## Build the Regression Model with sklean


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error

sk_reg = LinearRegression()

sk_reg.fit(X_train, y_train)

y_pred = sk_reg.predict(X_test)
'The mean square error is {0:.4f}'.format(mean_squared_error(y_test,y_pred))
```

## Normalize the Data 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## TensorFlow Model

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        # first layer has 5 neurons
        layers.Dense(5, activation='relu'),
        #last layer has one and only one neuron
        layers.Dense(1),
    ])

model.compile(optimizer='adam',loss='mse')
```

```python
tf.random.set_seed(1)
model.fit(x=X_train,y=y_train,batch_size=64,epochs=100,
          validation_data=(X_test,y_test)
          )
```

```python

y_pred = model.predict(X_test)
len(y_pred)
'The mean square error is {0:.4f}'.format(mean_squared_error(y_test,y_pred))
```

## Pytorch Models

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
```

```python
X_train = torch.tensor(X_train.astype(np.float32))
y_train = torch.tensor(y_train.values.astype(np.float32).reshape(-1,1))
y_train.shape
```

```python
input_size = X_train.shape[1]
output_size = y_train.shape[1]
#specify the number of neuron for the first layer
hidden_size = 10
print(input_size)
print(output_size)
```

```python
# define the child module class derivated from parent class of torch.nn.Module)
class LinearRegressionModel(torch.nn.Module):
    #define the constructor
    def __init__(self, input_size, hidden_size, output_size):
        super(LinearRegressionModel, self).__init__()
        self.hidden = torch.nn.Linear(input_size, hidden_size)  
        self.predict = torch.nn.Linear(hidden_size, output_size)  
    #overife the forward function in this child class
    def forward(self, x):
        x = F.relu(self.hidden(x))     
        y_pred = self.predict(x)            
        return y_pred
```

```python
model = LinearRegressionModel(input_size, hidden_size, output_size)
l = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)

#fix the random seeds for troch and np
torch.manual_seed(1)
np.random.seed(0)

#set the number of epochs
num_epochs = 100
for epoch in range(num_epochs):
    #forward pass
    y_pred = model(X_train.requires_grad_())

    #calculate the loss
    loss= l(y_pred, y_train)
    #Set the gradients to be zero
    optimizer.zero_grad()
    
    #backward pass: calculate gradients
    loss.backward()

    #update the weights
    optimizer.step()
  
    print('epoch {0}, loss:{1:.4f}'.format(epoch, loss.item()))
```

```python
#convert numpy to tensor
X_test = torch.from_numpy(X_test.astype(np.float32))
#Stop tracking the gradient by calling detach since we don't use it anymore
y_pred = model(X_test).detach().numpy()

'The mean square error is {0:.4f}'.format(mean_squared_error(y_test,y_pred))
```
